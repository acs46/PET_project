```python
from Bio import Entrez, Medline, SeqIO
Entrez.email = "astuart@ramapo.edu" 


handle = Entrez.esearch(db="pubmed", term="plastic AND degrading AND enzymes", retmax=50)
record = Entrez.read(handle)
idlist = record["IdList"]

print("Records Found:",record["Count"])

print("50 Id's for PubMed IDs of articles related to plastic and enzymes\n {}".format(idlist[:50]))

handle = Entrez.efetch(db="pubmed", id=idlist, rettype="medline")
records = Medline.parse(handle)
records = list(records)

for record in records:
    print("title:", record.get("TI", "?"))
    print("authors:", record.get("AU", "?"))
    print("source:", record.get("AB", "?"))
    print("")
```

```python
from Bio import Entrez
import time

try:
    from urllib.error import HTTPError  # for Python 3
except ImportError:
    from urllib2 import HTTPError  # for Python 2

Entrez.email = "astuart@ramapo.edu"

handle = Entrez.esearch(db="gds", term="Saccharomyces[kywd] AND beer[kywd]", RetMax='10', usehistory='y')
search_results = Entrez.read(handle)

print("The WebEnv is {}".format(search_results["WebEnv"]))
print("The QueryKey is {}".format(search_results["QueryKey"]))

print("Data Type of search_results ",type(search_results))
print(search_results.keys())
print("Results found =",search_results["Count"])
count = int(search_results["Count"])
print("Found %i results" % count)
print("Query Translation =",search_results["QueryTranslation"])
idlist = search_results["IdList"]
print(search_results["IdList"])


batch_size = 10

for start in range(0,count,batch_size):
    end = min(count, start+batch_size)
    print("Going to download record %i to %i" % (start+1, end))

    fetch_handle = Entrez.efetch(db="gds", retmode="text",
                                         retstart=start,
                                         retmax=10,
                                         #retmax=batch_size,
                                         webenv=search_results["WebEnv"],
                                         query_key=search_results["QueryKey"])
            
    for line in fetch_handle:
              print(line)
            
    #fetch_handle.close()
    
print("handle is type",type(fetch_handle))
```
